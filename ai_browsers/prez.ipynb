## Use Cases

AI browsers excel at automating manual workflows that span multiple tools and tabs. Here are real-world examples:

### Compliance & Documentation
- **Compliance questionnaires**: Reduced from 7-8 hours to 3.5 hours with final human validation
- **Receipt compilation**: Automated gathering from Gmail, spreadsheets, and vendor sites for accounting

### Workflow Automation  
- **Task migration**: Automatically compile Slack conversations into structured Notion tickets
- **Location mapping**: Determine rail locations using Google Maps and technical diagrams (saves 20 min/task × thousands annually)

### When to Use AI Browsers
Perfect for repetitive tasks requiring:
- Navigation across multiple websites/applications
- Data gathering and compilation
- Pattern-based decision making

### Best Practices for Eval Dataset Size
- Quality > Quantity for complex enterprise tasks
- **100-200 high-quality examples** typically sufficient for reliable performance signals
- Applies to: summarization, document extraction, and similar complex workflows
- Focus on curated, representative samples over massive datasets
